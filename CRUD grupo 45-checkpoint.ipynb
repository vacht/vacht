```python
from datetime import datetime
from openpyxl import load_workbook

def leer(ruta:str, extraer:str):#<--- creamos una funcion que recibe como parametro la ruta del documento Exccel y el estado dela tarea que queremos buscar
    Archivo_Exccel = load_workbook(ruta) #<-- gruardamos en arvhivo exccel espesificado con la ruta en una variable 
    Hoja_datos = Archivo_Exccel['Datos del crud'] #<-- seleccionamos la hoja del archivo excel que queremos leer
    Hoja_datos=Hoja_datos['A2':'F'+str(Hoja_datos.max_row)]#<--- extraigo la informacion de la hoja desde A2 hasta la maxima fila de F
    info={}#<-- creo un diccionario donde organizare los datos leidos desde el excel
    
    for i in Hoja_datos:#<-- recorremos la variable Hoja_datos, lacual es una tupla, la llaves de esta tubla son la filas del documento Execcl y el valor de estas llaves son una tupla con los datos de las columnas de la fila
        #i toma el valor de las las tuplas, cullo indices 1, 2, 3 etc, equivale a la columna A, B, C etc en el documento Exccel
        if isinstance(i[0].value, int):
            info.setdefault(i[0].value,{'tarea':i[1].value, 'descripcion':i[2].value, 'estado':i[3].value, 'fecha de inicio':i[4].value, 'fecha de finalizacion':i[5].value })#<--usamos el atributo .value para ir tomando el valor de la celda y guardarla en el diccionario
        # el diccionario info guarda toda la informacion de archivo Exccel
    
    if not(extraer=='todo'):#<-- comprobamos que filtro queremos aplicar si es  "todo" no llamamos a la funcion filtro, de lo controrio llamamos a la fincion filtro
        info=filtrar(info, extraer)#<-- llamada a la funcion filtro
    
    for i in info:#<- una vez filtrado el diccionario imprimimos en consola el cada una de las tareas 
        print('******** Tarea *******')
        print('Id:'+ str(i)+'\n'+'Titulo: '+str(info[i]['tarea'])+'\n'+'Descripcion: '+str(info[i]['descripcion']) + '\n'+ 'Estado:'+str(info[i]['estado']) + '\n'+'Fecha Creacion: '+ str(info[i]['fecha de inicio']) + '\n'+'Fecha de finalizacion: ' +str(info[i]['fecha de finalizacion']) )
        print()
        
    return 


def filtrar(info:dict,filtro:str):#<-- esta funcion recibe como parametro el diccionario con toda la informacion del documento Exccel y el estado por el cual queremos filtrar
    aux={}#<-- creamos un diccionario auxiliar que guardara las tareas que cumplan con el filtro

    for i in info: #<-- recorremos el diccionario info
        if info[i]['estado']==filtro: #<-- preguntamos si el estado de la tarea es igual al filtro que deseamos consultar
            aux.setdefault(i,info[i])#<- si cumple con la condicion grardamos i como la llave y info[i] colo el contenido, para que aux mantenta la misma estructura de diccionario info
    return aux
    
def actualizar(ruta:str, identificador:int,datos_actualizados:dict):#<-- funcion que recibe como parametros la ruta del archivo, el Id de la tarea que editaremos y un diccionario con los datos que editaremos
    Archivo_Exccel = load_workbook(ruta) #<-- gruardamos en arvhivo exccel espesificado con la ruta en una variable 
    Hoja_datos = Archivo_Exccel['Datos del crud'] #<-- seleccionamos la hoja del archivo excel que queremos leer
    Hoja_datos=Hoja_datos['A2':'F'+str(Hoja_datos.max_row)]#<--- extraigo la informacion de la hoja desde A2 hasta la maxima fila de F
    hoja=Archivo_Exccel.active#<-- del archivo exccel selecciono la primera hora 
    #a diferencia de las listas cullos indices comienzan en 0, los archivos excel su indice comienza en 1 por lo tanto la columna A equivale a indice 1 la B al indice 2 y asi con todas las filas del documento
    
    titulo=2 # <-al conocer la estructura del archivo exccel sabemos que la celda del titulo esta en la fila B que corresponde al indice 2
    descripcion=3 #<- la celda de descripcion esta en la columna C correspondiente al indice  3
    estado=4#<-- estado esta en la columna D correspondiente al indice 5
    fecha_inicio=5
    fecha_FInalizado=6
    encontro=False#<-- usamos una variable booleana para saber si se encontro el id que se espesifico, este tipo de varibles tambien es conocido como banderas
    for i in Hoja_datos: #<- recorremos la hoha de datos
        if i[0].value==identificador:#<-- preguntamos si el id que estamos leyendo es igual al que estamos buscando
            fila=i[0].row #<-- el atributo .row es similar al atributo .value, con la diferencia que este nos revuelve la fila en donde se encuentra 
            encontro=True#<-- colocamos la bandera de encontrado en true
            for d in datos_actualizados: #<-- una vez encontrado el id conicido en que fila esta, recorremos el vector de datos actualizados, y vmos agregando los nuevos datos
                if d=='titulo' and not(datosActualizados[d]==''): #<-- si el la llave del diccionarioActualizados es titulo y el valor de la llave no es vacio entonces actualizadmos
                    hoja.cell(row=fila, column=titulo).value=datosActualizados[d] #<- usamos el atributo .cell el cual recibe los parametros fila y conlumna cullos atributos ya sabemos
                elif d=='descripcion' and not(datosActualizados[d]==''):
                    hoja.cell(row=fila, column=descripcion).value=datosActualizados[d]
                elif d=='estado' and not(datosActualizados[d]==''):
                    hoja.cell(row=fila, column=estado).value=datosActualizados[d]
                elif d=='fecha inicio' and not(datosActualizados[d]==''):
                    hoja.cell(row=fila, column=fecha_inicio).value=datosActualizados[d]
                elif d=='fecha finalizacion' and not(datosActualizados[d]==''):
                    hoja.cell(row=fila, column=fecha_FInalizado).value=datosActualizados[d]       
    Archivo_Exccel.save(ruta) #<--- guardamos los cambios del archivo exccel
    if encontro==False:
        print('Error: No existe una tarea con ese Id')#<- en caso que no se encuetre el imprime que no seconsigio una tarea con el id
        print()
    return

def agregar(ruta:int, datos:dict):#< funcion que recibe la direccion del archivo y un diccionario con los datos que deseamos agregar
    Archivo_Exccel = load_workbook(ruta) #<-- gruardamos en arvhivo exccel espesificado con la ruta en una variable 
    Hoja_datos = Archivo_Exccel['Datos del crud'] #<-- seleccionamos la hoja del archivo excel que queremos leer
    Hoja_datos=Hoja_datos['A2':'F'+str(Hoja_datos.max_row+1)]#<--- extraigo la informacion de la hoja desde A2 hasta la maxima fila de F
    hoja=Archivo_Exccel.active#<-- del archivo exccel selecciono la primera hora 
    
    titulo=2#<-- igual que en actualizar selecionamos guardamos los indices
    descripcion=3
    estado=4
    fecha_inicio=5
    fecha_FInalizado=6
    for i in Hoja_datos:#<-- recorremos el archivo 
        
        if not( isinstance(i[0].value, int)): #<- como sabemos que el id de las tareas es un entero, usamos isinstance para preguntar si es un entero, ya que si no es entero quiere decir que la fila esta libre y ahi agregaremos nuestra nueva tarea
            identificador=i[0].row  #<-- usamos nuevamente el atributo .row para saber la fila libre y prosedemos agregar los datos
            hoja.cell(row=identificador, column=1).value=identificador-1
            hoja.cell(row=identificador, column=titulo).value=datos['titulo']
            hoja.cell(row=identificador, column=descripcion).value=datos['descripcion']
            hoja.cell(row=identificador, column=estado).value=datos['estado']
            hoja.cell(row=identificador, column=fecha_inicio).value=datos['fecha inicio']       
            hoja.cell(row=identificador, column=fecha_FInalizado).value=datos['fecha finalizacion']       
            break
    Archivo_Exccel.save(ruta)
    return

def borrar(ruta,identificador):
    Archivo_Exccel = load_workbook(ruta) #<-- gruardamos en arvhivo exccel espesificado con la ruta en una variable 
    Hoja_datos = Archivo_Exccel['Datos del crud'] #<-- seleccionamos la hoja del archivo excel que queremos leer
    Hoja_datos=Hoja_datos['A2':'F'+str(Hoja_datos.max_row)]#<--- extraigo la informacion de la hoja desde A2 hasta la maxima fila de F
    hoja=Archivo_Exccel.active#<-- del archivo exccel selecciono la primera hora 
    #a diferencia de las listas cullos indices comienzan en 0, los archivos excel su indice comienza en 1 por lo tanto la columna A equivale a indice 1 la B al indice 2 y asi con todas las filas del documento
    
    titulo=2 # <-al conocer la estructura del archivo exccel sabemos que la celda del titulo esta en la fila B que corresponde al indice 2
    descripcion=3 #<- la celda de descripcion esta en la columna C correspondiente al indice  3
    estado=4#<-- estado esta en la columna D correspondiente al indice 5
    fecha_inicio=5
    fecha_FInalizado=6
    encontro=False#<-- usamos una variable booleana para saber si se encontro el id que se espesifico, este tipo de varibles tambien es conocido como banderas
    for i in Hoja_datos: #<- recorremos la hoha de datos
        if i[0].value==identificador:#<-- preguntamos si el id que estamos leyendo es igual al que estamos buscando
            fila=i[0].row #<-- el atributo .row es similar al atributo .value, con la diferencia que este nos revuelve la fila en donde se encuentra 
            encontro=True#<-- colocamos la bandera de encontrado en true
            
            hoja.cell(row=fila, column=1).value=""
            hoja.cell(row=fila, column=titulo).value=""
            hoja.cell(row=fila, column=descripcion).value=""
            hoja.cell(row=fila, column=estado).value=""
            hoja.cell(row=fila, column=fecha_inicio).value=""   
            hoja.cell(row=fila, column=fecha_FInalizado).value=""  
    Archivo_Exccel.save(ruta) #<--- guardamos los cambios del archivo exccel
    if encontro==False:
        print('Error: No existe una tarea con ese Id')#<- en caso que no se encuetre el imprime que no seconsigio una tarea con el id
        print()
    return



rut=r"C:\Users\freck\OneDrive\Escritorio\crud45.xlsx"   

datosActualizados={'titulo':'', 'descripcion':'', 'estado':'', 'fecha inicio':'', 'fecha finalizacion':''}
while True:
    print('Indique la accion que desea realizar: ')
    print('Consultar: 1')
    print('Actualizar: 2')
    print('Crear nueva tarea: 3')
    print('Borrar: 4')
    accion = input('Escriba la opcion:')
    if not(accion=='1') and not(accion=='2') and not(accion=='3') and not(accion=='4'):
        print('Comando invalido por favor eliga una opcion valida')
    elif accion=='1':
        opc_consulta=''
        print('Indique la tarea que desea consultar: ')
        print('Todas las tareas: 1')
        print('En espera: 2')
        print('En ejecucion: 3')
        print('Por aprobar: 4')
        print('Finalizada: 5')
        opc_consulta = input('Escriba la tarea que dese consultar:')
        if opc_consulta=='1':
            print()
            print()
            print('** Consultando todas las tareas **')
            leer(rut,'todo')
    elif opc_consulta=='2':
            print()
            print()
            print('** Consultando tareas en espera **')
            leer(rut,'En espera')
    elif opc_consulta=='3':
            print()
            print()
            print('** Consultando tareas en ejecucion **')
            leer(rut,'En ejecucion')
    elif opc_consulta=='4':
            print()
            print()
            print('** Consultando tareas por aprobar **')
            leer(rut,'Por aprobar')
            
    elif opc_consulta=='5':
            print()
            print()
            print('** Consultando tareas finalizadas **')
            leer(rut,'Finalizada')
    elif accion=='2':
        datosActualizados={'titulo':'', 'descripcion':'', 'estado':'', 'fecha inicio':'', 'fecha finalizacion':''}
        print('** Actualizar Tarea **')
        print()
        id_Actualizar=int(input('Indique el Id de la tarea que desea actualizar: '))
        print()
        print('** Nuevo titulo **')
        print('**Nota: si no desea actualizar el titulo solo oprima ENTER')
        datosActualizados['titulo']=input('Indique el nuevo titulo de la tarea : ')
        print()
        print('** Nueva descripcion **')
        print('**Nota: si no desea actualizar la descripcion solo oprima ENTER')
        datosActualizados['descripcion']= input('Indique la nueva descripcion de la tarea : ')
        print()
        print('** Nueva estado **')
        print('En espera: 2')
        print('En ejecucion: 3')
        print('Por aprobar 4')
        print('Finalizada: 5')
        
        print('**Nota: si no desea actualizar el estado solo oprima ENTER')
        estadoNuevo= input('Indique el nuevo estado de la tarea : ')
        if estadoNuevo=='2':
            datosActualizados['estado']='En espera'
        elif estadoNuevo=='3':
            datosActualizados['estado']='En ejecucion'
        elif estadoNuevo=='4':
            datosActualizados['estado']='Por aprobar'
        elif estadoNuevo=='5':
            now = datetime.now()
            datosActualizados['estado']='Finalizada'
            datosActualizados['fecha finalizacion']=str(now.day) +'/'+ str(now.month) +'/'+str(now.year) 
        
        now = datetime.now()
        datosActualizados['fecha inicio']=str(now.day) +'/'+ str(now.month) +'/'+str(now.year) 
        actualizar(rut,id_Actualizar, datosActualizados)
        print()
    elif accion=='3':
        datosActualizados={'tarea':'', 'descripcion':'', 'estado':'', 'fecha inicio':'', 'fecha finalizacion':''}
        print('** Crear nueva Tarea **')
        
        print()
        print('** titulo **')
        print()
        datosActualizados['titulo']=input('Indique el titulo de la tarea : ')
        print()
        print('** descripcion **')
        datosActualizados['descripcion']= input('Indique la descripcion de la tarea : ')
        print()
        datosActualizados['estado']='En espera'
        now = datetime.now()
        datosActualizados['fecha inicio']=str(now.day) +'/'+ str(now.month) +'/'+str(now.year)
        datosActualizados['fecha finalizacion']=''
        agregar(rut,datosActualizados)
    elif accion=='4':
        print('')
        print('** Eliminar Tarea **')
        iden=int(input('Indique el Id de la tarea que desea eliminar: '))
        borrar(rut,iden)
```
